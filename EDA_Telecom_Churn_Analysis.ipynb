## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Finding out the number of customers who are opting out of the subscription.
churns_number= (df["Churn"]==True).sum()
print(f'The number of customers Churns are :{churns_number}')
#Finidng the number of customers retained or active subscribers.
staying_number = (df["Churn"]==False).sum()
print(f'The number of customer Staying are :{staying_number}')
````

````python
#State wise Number of customer churns.
State_Churn=df[["State","Churn"]].reset_index()
State_Churn= State_Churn.groupby("State")["Churn"].value_counts().reset_index(name='Churn_Number')
State_Churn['Churn_percent']=round(State_Churn["Churn_Number"]/State_Churn['Churn'].sum()*100,2)
State_Churn[State_Churn['Churn']==True].sort_values(by='Churn_percent',ascending=False).head(20).reset_index(drop=True)
````

````python
#Area code wise Number of customer churns.
Ac_Churn=df[["Area code","Churn"]]
Ac_Churn= Ac_Churn.groupby("Area code")["Churn"].value_counts().reset_index(name='Count')
Ac_Churn
````

````python
#Churns per Area code
Ac_Churn[Ac_Churn['Churn']==True]
````

````python
#Analysing the percentage of churns ratio to the customers reatining based on the International Plan.
Int_plan_churn=pd.crosstab(df["International plan"],df["Churn"])
Int_plan_churn["Percent_churn"]=round(Int_plan_churn[True]/(Int_plan_churn[True]+Int_plan_churn[False])*100)
Int_plan_churn
````

````python
#analysing the customer churns based on the Voicemail Plan subscription 
Voicemail_Plan_data=pd.crosstab(df["Voice mail plan"],df["Churn"])
Voicemail_Plan_data["Percentage_of_Churns"]=(Voicemail_Plan_data[True]/(Voicemail_Plan_data[True]+Voicemail_Plan_data[False]))*100
Voicemail_Plan_data
````

````python

# Analysing the customer churns based on the Number of Voicemail Messages.
vm_yes=df[["Voice mail plan",'Number vmail messages','Churn']]
vm_yes[(vm_yes['Number vmail messages'])& (vm_yes['Churn']==True)].sort_values(by='Number vmail messages',ascending=False).reset_index(drop=True)
````

````python
#Total sum of the calls made during different sections of the Day and Calculating the Total Call Costs.
calls_data=pd.DataFrame(columns=["total day minutes","total day charge","total eve minutes","total eve charge","total night minutes","total night charge"])
calls_data["total day minutes"]=pd.Series(df["Total day minutes"].sum())
calls_data["total eve minutes"]=pd.Series(df["Total eve minutes"].sum())
calls_data["total night minutes"]=pd.Series(df["Total night minutes"].sum())
calls_data["Total intl minutes"]=pd.Series(df["Total intl minutes"].sum())
calls_data["total day charge"]=pd.Series(df["Total day charge"].sum())
calls_data["total eve charge"]=pd.Series(df["Total eve charge"].sum())
calls_data["total night charge"]=pd.Series(df["Total night charge"].sum())
calls_data["total international call charge"]=pd.Series(df["Total intl charge"].sum())
calls_data 
````

````python
#Now calculating the Per Minute cost for calls made during different sessions of the Day.
calls_data.insert(2,"PM_DayCharge",round(calls_data["total day charge"]/calls_data["total day minutes"],2))
calls_data.insert(5,"PM_Eve_Charge",round(calls_data["total eve charge"]/calls_data["total eve minutes"],2))
calls_data.insert(8,"PM_Night_Charge",round(calls_data["total night charge"]/calls_data["total night minutes"],2))
calls_data.insert(11,"PM_Int_Charge",round(calls_data["total international call charge"]/calls_data["Total intl minutes"],2))	
calls_data
````

````python
#Checking for cost variation ito the customers with and without International plan subscription
Int_cal_cost=df[['International plan','Total intl minutes','Total intl charge']]
Int_cal_cost['Cost_per_min']=Int_cal_cost['Total intl charge']/Int_cal_cost['Total intl minutes']
Int_cal_cost.groupby('International plan')['Cost_per_min'].mean().reset_index(name='Cost_per_min')
````

````python
# Analysing Customer service calls by calculating the percentage of Churns.
Cusc_data=pd.crosstab(df["Customer service calls"],df["Churn"])
Cusc_data["Churn_percentage"]=round(Cusc_data[True]*100/(Cusc_data[True]+Cusc_data[False]),2)
Cusc_data
````

````python
#Analysing the Area code,International plan, Voicemail plan,Churn features for getting better insights.
data=df[['Area code','International plan', 'Voice mail plan','Churn']]
data = data[data["Churn"]==True].reset_index(drop=True)
data
````

````python
#Finding the number of churns per each Area Code who are subscribed to both internatiional and voicemail plans.
data_Inp_Vmp_yes=data[(data['International plan']=='Yes')&(data['Voice mail plan']=='Yes')]
#printing the shape of churned customers with subscription to International plan and Voicemail Plan
print("Total number of customers churned who have subscription to both International plan and Voicemail Plan are",data_Inp_Vmp_yes.shape[0],end='\n')
print(" ")
data_Inp_Vmp_yes=data_Inp_Vmp_yes.groupby('Area code')["Churn"].value_counts().reset_index(name='Count')

#printing the Area code wise customer churns
index=0
for i in data_Inp_Vmp_yes['Area code']:
  print(f"For Area code {i},","the number of customer churns who subscribed to both International Plan and Voicemail Plan are", data_Inp_Vmp_yes['Count'][index],end="\n")
  index+=1
  print("")
````

````python
#Finding the customers who are churned with International plan subscription but don't have Voicemail plan subscription.
data_Inp_yes_Vmp_no=data[(data['International plan']=='Yes')&(data['Voice mail plan']=='No')]
#printing the number of churned customers with subscription to International plan but not for Voicemail Plan.
print("Total number of customers churned who have subscription to International plan but not for Voicemail Plan are: ",data_Inp_yes_Vmp_no.shape[0],end='\n')
print(" ")
data_Inp_yes_Vmp_no=data_Inp_yes_Vmp_no.groupby('Area code')["Churn"].value_counts().reset_index(name='Count')

#printing the Area code wise customer churns
index=0
for i in data_Inp_yes_Vmp_no['Area code']:
  print(f"For Area code {i},","the number of customers churned who have subscription to International plan but not for Voicemail Plan are: ", data_Inp_yes_Vmp_no['Count'][index],end="\n")
  index+=1
  print("")
````

````python
#Finding the customers who are churned with No International plan subscription but have Voicemail plan subscription.
data_Inp_No_Vmp_Yes=data[(data['International plan']=='No')&(data['Voice mail plan']=='Yes')]
#printing the shape of churned customers with subscription to Voicemail Plan but not for International plan.
print("Total number of customers churned who have subscription to Voicemail Plan but not for International plan are: ",data_Inp_No_Vmp_Yes.shape[0],end='\n')
print(" ")
data_Inp_No_Vmp_Yes=data_Inp_No_Vmp_Yes.groupby('Area code')["Churn"].value_counts().reset_index(name='Count')

#printing the Area code wise customer churns
index=0
for i in data_Inp_No_Vmp_Yes['Area code']:
  print(f"For Area code {i},","the number of customers churned who have subscription to Voicemail Plan but not for International plan are: ", data_Inp_No_Vmp_Yes['Count'][index],end="\n")
  index+=1
  print("")
````

````python
#Finding the customers who are churned with No subscription for International plan and Voicemail plan.
data_Inp_Vmp_No=data[(data['International plan']=='No')&(data['Voice mail plan']=='No')]
#printing the shape of churned customers with No subscription for International plan and Voicemail plan.
print("Total number of customers churned who don't have subscription for International plan and Voicemail plan are: ",data_Inp_Vmp_No.shape[0],end='\n')
print(" ")

data_Inp_Vmp_No=data_Inp_Vmp_No.groupby('Area code')["Churn"].value_counts().reset_index(name='Count')

#printing the Area code wise customer churns
index=0
for i in data_Inp_Vmp_No['Area code']:
  print(f"For Area code {i},","the number of customers churned who don't have subscription for International plan and Voicemail plan are: ", data_Inp_Vmp_No['Count'][index],end="\n")
  index+=1
  print("")
````

````python
#finding the Top 10 States with Higher number of churns.

states_data=df[["State","Churn"]]
states_data= states_data[states_data["Churn"]==True]
Sc_plot=states_data.groupby("State")["Churn"].sum().head(10).sort_values(ascending=False)
Sc_plot.plot.bar(figsize=(15,6))
plt.title("Top 10 States with highest number of Churns",fontsize=15)
plt.ylabel("Number of Churns",fontsize=12)
plt.show()
````
